```python
from diffusers import StableDiffusionXLPipeline
import torch
import matplotlib.pyplot as plt
```

```python
# Parametros do nosso pipeline
pretrained_model ="stabilityai/stable-diffusion-xl-base-1.0"
prompt = 'A colorful vibtreerant city st, moonlight, a red car parked by the sidewall'
negative_prompt = "blurry, low quality , grayscale, monochrome, dull, dark" # Oq nao queremos

num_inference_steps = 70 #Numero de etapas
guidance_scale =  7.5
height = 768
width = 768
seed = 42
```

```python
generator = torch.manual_seed(seed)

pipeline = StableDiffusionXLPipeline.from_pretrained(
    pretrained_model,
    torch_dtype = torch.float16,
    use_safetensors = True,


).to("cuda")
```

```python
image = pipeline(
    prompt = prompt,
    negative_prompt = negative_prompt,
    width= width,
    height=height,
    num_inference_steps= num_inference_steps,
    guidance_scale= guidance_scale,
    generator= generator
).images[0]
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```
