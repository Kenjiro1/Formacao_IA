```python
import pandas as pd
from keras.utils import plot_model
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout #Dense é a full conected, dropout é a técnica de regularização
```

```python
url_churn = 'https://raw.githubusercontent.com/Kenjiro1/Formacao_IA/refs/heads/main/dados/Churn_treino.csv'


df = pd.read_csv(url_churn, sep=';')
df.head()
```

```python
X = df.drop('Exited',axis=1)
y = df['Exited']
```

#Etapa de padronização dos dados


```python
standardscaler = StandardScaler()
numerical = X.select_dtypes(include =['int64','float64']).columns

X[numerical]= standardscaler.fit_transform(X[numerical])
```

```python
X
```

#Label Encoder


```python
labelencoder = LabelEncoder()
categorical = X.select_dtypes(include='object').columns
for col in categorical:
  X[col] = labelencoder.fit_transform(X[col])


X
```

```python
X_treino, X_teste, Y_treino,Y_teste = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

#Modelo Sequential

```python
modelo = Sequential()
modelo.add(Dense(units=64, activation ='relu', input_dim=X_treino.shape[1])) #Units = numero neuronios, input_dim = tamanho da entrada
modelo.add(Dropout(0.4))
modelo.add(Dense(units = 32, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units = 64, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units = 1, activation ='sigmoid'))
 #Saida é 0 ou 1 ou seja ou ele ativa ou não por tanto um neuronio
 #Sigmoid como função de ativação retorna uma probabilidade o que é bem interessante
```

```python
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
modelo.fit(X_treino,Y_treino, epochs= 50, batch_size = 32)
```

```python
previsao = modelo.predict(X_teste)
previsao
```

```python
y_pred = (previsao > 0.5).astype('int32')
```

```python
print("Acuracia " , accuracy_score(Y_teste,y_pred))
print("F1 " , f1_score(Y_teste,y_pred))
print("Recall " , recall_score(Y_teste,y_pred))
print("Confusion Matrix " , confusion_matrix(Y_teste,y_pred))
```
