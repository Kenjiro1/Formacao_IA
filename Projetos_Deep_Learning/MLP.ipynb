```python
#Multi layer perceptron
#Vamo utilizar o iris dataset

from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X_treinamento,X_teste,Y_treinamento,Y_teste = train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)
```

```python
modelo = MLPClassifier(
    verbose=True, #gera um output a cada iteração
    hidden_layer_sizes=(5,4), #(5,4) Duas camadas oculdas uma com 5 neuronios e outra com 4
    activation='relu',
    batch_size=20,
    learning_rate='adaptive',
    momentum=0.9,
    early_stopping=False,
    max_iter= 1000,
    random_state=10#Epochs
)


modelo.fit(X_treinamento, Y_treinamento)
```

```python
#Grafico de Loss
plt.plot(modelo.loss_curve_)
plt.xlabel('Iterações')
plt.ylabel('Valor da loss')
plt.title('loss')
plt.show()
```

```python
previsoes = modelo.predict(X_teste)

accuracy_score(Y_teste, previsoes)
```
