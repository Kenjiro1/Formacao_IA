```python
import pygad
import numpy as np
```

```python
itens = [
    ("canivete",10,1),
    ("feijão",20,5),
    ("batatas",15, 10),
    ("lanterna",2,1),
    ("saco de dormir", 30, 7),
    ("corda",10,5),
    ("bussula",30,1)
    ]


limite_peso = 15
```

```python
def fitness_func(ga_instance, solution, solution_idx):
  pontos = 0
  peso = 0
  for i in range(len(solution)):
    if solution[i] == 1: # Item esta incluido
      pontos +=itens[i][1]
      peso += itens[i][2]
  if peso > limite_peso:
    return 1
  return pontos



ga_instance = pygad.GA(
      num_generations= 200,
      num_parents_mating= 5,
      fitness_func= fitness_func,
      sol_per_pop = 50,
      num_genes = len(itens), # Gene eh o numero de itens que temos
      gene_type = int,
      init_range_high= 2, # Na pratica 1
      init_range_low=0

  )



ga_instance.run()
solution, solution_fitness,_ = ga_instance.best_solution()
print('Melhor solucao', solution)
print('Melhor solucao', solution_fitness)
```
