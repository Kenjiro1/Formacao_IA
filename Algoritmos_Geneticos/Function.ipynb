```python
# função de adaptação
#2 + x + 5 = 20
import numpy as np
import matplotlib.pyplot as plt

def equation_func(solution): # Solution = solução proposta pelo algoritmo genetico
  X = solution[0]
  equation_output = 2*X**2 + 5
  target_output = 20 # Nossa solução
  fitness = np.abs(equation_output - target_output + 1e-8)
  return fitness
```

```python
lista = [-10,-5,0,5,7.5,10,15,20,25]
solucoes = []
for value in lista:
  result = equation_func([value])
  solucoes.append(result)
```

```python
min(solucoes)
```

```python
plt.plot(lista, solucoes)
plt.plot(min(solucoes), 'ro', label = min(solucoes))
plt.legend()
plt.xlabel('Solucao proposta')
plt.ylabel('Retorno da funcao de ativacao')
plt.show()
```
