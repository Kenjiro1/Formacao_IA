```python
import pygad
import numpy as np
```

```python
#Equacao function
def equation_func(ga_instance, solution, solution_idx): # Solution = solução proposta pelo algoritmo genetico
  X = solution[0]
  equation_output = 2*X**2 + 5
  target_output = 20 # Nossa solução
  fitness = np.abs(equation_output - target_output + 1e-8)
  return fitness
```

```python
ga_instance = pygad.GA(
    num_parents_mating = 10,
    num_generations= 50,
    fitness_func = equation_func,
    sol_per_pop = 20,
    num_genes=1,
    gene_type = float,
    init_range_low= -10,
    init_range_high= 10
    #mutation_percent_genes = 30
    #mutation_type =None
)


ga_instance.run()
solution, solution_fitness , _ = ga_instance.best_solution()
print('Melhor solucao ', round(solution[0], 1))
print('Pontos:', solution_fitness)
```

```python
#grafico da convergencia
ga_instance.plot_fitness()
```
