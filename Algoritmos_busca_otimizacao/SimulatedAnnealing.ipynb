```python
import numpy as np
from scipy.optimize import basinhopping
```

```python
def rosenbrock(x):
  return (1 - x[0]) **2 + 100*(x[1] - x[0]**2)**2
```

```python
#limites da busca
bounds = [(-2,2), (-2,2)]
x0 = np.array([0,0])

options={'stepsize' : 0.5 , 'T' : 1.0, 'niter': 100,'disp': False} #Step size: passos, T: temperatura ,
```

```python
result = basinhopping(rosenbrock, x0, minimizer_kwargs={'bounds' : bounds}, **options ) #**dicionario como parametro
```

```python
print(result)
```
