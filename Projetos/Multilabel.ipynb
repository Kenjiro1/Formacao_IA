```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import hamming_loss
from sklearn.multioutput import ClassifierChain
from sklearn.preprocessing import StandardScaler
```

```python
musica = pd.read_csv('https://raw.githubusercontent.com/Kenjiro1/Formacao_IA/refs/heads/main/dados/Musica.csv')
```

```python
musica.head()
```

```python
musica.shape
```

```python
#Vamos separar as classes das variaveis independentes

classe = musica.iloc[:,0:6].values
previsores = musica.iloc[:,7:78].values
```

```python
scaler = StandardScaler()
#Padronizamos
previsores = scaler.fit_transform(previsores)
#Nao eh necessario fazer isso com a classe por ser uma classe
```

```python
x_train, x_teste, y_train, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state =5)
```

```python
#Vamos criar o modelo
svm = SVC()
classifier_chain = ClassifierChain(base_estimator = svm, random_state =5)
classifier_chain.fit(x_train,y_train)

```

```python
y_pred = classifier_chain.predict(x_teste)

loss= hamming_loss(y_teste, y_pred)

print(f"Hamming loss: {loss}" )  #18% de erro
```
