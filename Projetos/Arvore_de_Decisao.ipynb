```python
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import export_graphviz
import graphviz
```

```python
import math
from collections import Counter

def calcular_entropia(dados):
    """
    Calcula a entropia de uma lista de rótulos (classes).

    Parâmetros:
    dados (list): Lista de rótulos (por exemplo, ['sim', 'não', 'sim', 'sim'])

    Retorna:
    float: Valor da entropia
    """
    total = len(dados)
    contagem = Counter(dados)

    entropia = 0.0
    for rotulo, freq in contagem.items():
        p = freq / total
        entropia -= p * math.log2(p)

    return entropia

# Exemplo de uso
rotulos = ['sim', 'não', 'sim', 'sim', 'não']
print("Entropia:", calcular_entropia(rotulos))
```

#Curso

```python
CSV_GITHUB_URL = "https://raw.githubusercontent.com/Kenjiro1/Formacao_IA/refs/heads/main/dados/insurance.csv"
base = pd.read_csv(CSV_GITHUB_URL, keep_default_na=False)
base.head()
```

```python
base.isnull().sum()
```

```python
#Our Target it is accident
Y = base['Accident']
X = base.drop(columns = ['Accident'],axis=1).values
print(type(X))
print("="* 20)
print(type(Y))
print("="* 20)
X
```

```python
encoder = LabelEncoder()
for i in range(X.shape[1]):
  if X[:,i].dtype == 'object':
    X[:,i] = encoder.fit_transform(X[:,i])

X
```

```python
X_treinamento, X_test, Y_treinamento, Y_test = train_test_split(X,Y,test_size=0.3, random_state=12)

# Do a loop that says the best Max_depth and max_leaf_nodes
modelo = DecisionTreeClassifier(random_state = 1, max_depth=8, max_leaf_nodes=6)
modelo.fit(X_treinamento, Y_treinamento)
```

```python
dot_data = export_graphviz(modelo, out_file=None, filled = True,feature_names=base.columns[:-1],
                           class_names=True, rounded = True)
graph = graphviz.Source(dot_data)
graph.render("DecisionTree", format ="png")
graph.view()
```

```python
predict = modelo.predict(X_test)
print(classification_report(Y_test, predict)) #Metrics based on each class
```

```python
print(accuracy_score(Y_test, predict))
print(f1_score(Y_test, predict, average='weighted'))
print(precision_score(Y_test, predict, average='weighted'))
print(recall_score(Y_test, predict, average='weighted'))
```
