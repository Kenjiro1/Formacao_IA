```python
def plot_cluster(data,labels, title):
  colors = ['red','green','purple','black']
  plt.figure(figsize = (4,8))
  for i,c,l in zip(range(-1,3),colors,['Noise','Setosa','Versicolor','Virginica']):
    if i == -1:
      plt.scatter(data[labels==i,0],data[labels==i,1],c=colors[i],label=l, alpha= 0.25, s=50, marker ='x')
    else:
        plt.scatter(data[labels==i,0],data[labels==i,1],c=colors[i],label=l, alpha= 0.25, s=50)

  plt.legend()
  plt.title(title)
  plt.xlabel('Comprimento da petala')
  plt.ylabel('Largura da petala')
  plt.show()
```

```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram


```

```python
iris = datasets.load_iris()
iris
```

```python
iris.target
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(iris.data)
print(kmeans.labels_)
```

```python
confusion_matrix(iris.target, kmeans.labels_)


plot_cluster(iris.data,kmeans.labels_,'K-Means')
```

```python
#DBSCAN pode gerar ruido
dbscan = DBSCAN(eps=0.5, min_samples=3)
dbscan.fit(iris.data)
print(dbscan.labels_)
confusion_matrix(iris.target, dbscan.labels_)

plot_cluster(iris.data,dbscan.labels_,'DBSCAN')
```

```python
agglo = AgglomerativeClustering(n_clusters=3)
agglo.fit(iris.data)
print(agglo.labels_)

confusion_matrix(iris.target, agglo.labels_)

plot_cluster(iris.data,agglo.labels_,'Agrupamento hierarquico')
```
